```python
# Import the required libraries and dependencies
import os
import requests
import json
import pandas as pd
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi
from MCForecastTools import MCSimulation
from alpaca_trade_api import TimeFrame

%matplotlib inline

load_dotenv()
```

```python
# Set the variables for the Alpaca API and secret keys
alpaca_api_key = os.getenv('Alpaca_API_Key')
alpaca_secret_key = os.getenv('Alpaca_Secret_Key')

display(type(alpaca_api_key))
display(type(alpaca_secret_key))

# Create the Alpaca tradeapi.REST object
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version='V2')
```

```python
# Set the tickers for both the bond and stock portion of the portfolio
portfolio_tickers = ['GME','AAPL','MSFT','AMZN']
comparison_tickers = ['SPY','AGG']

# Set timeframe to 1D 
timeframe = TimeFrame.Day

number_of_trading_days = 252
number_of_tickers = len(tickers)
limit = number_of_trading_days*number_of_tickers

limit_rows = 5000

# Format current date as ISO format
# Set both the start and end date at the date of your prior weekday 
# This will give you the closing price of the previous trading day
# Alternatively you can use a start and end date of 2020-08-07
start_date = pd.Timestamp('2018-03-18', tz='America/New_York').isoformat()
end_date = pd.Timestamp('2022-03-18', tz='America/New_York').isoformat()
limit
```

```python
# Use the Alpaca get_barset function to get current closing prices the portfolio
# Be sure to set the `df` property after the function to format the response object as a DataFrame
prices_df = alpaca.get_bars(tickers, timeframe ,start=start_date, end=end_date, limit=limit_rows).df

# Review the first 5 rows of the Alpaca DataFrame
prices_df.head()
```

```python
prices_ticker_1_df = alpaca.get_bars(tickers[0], timeframe ,start=start_date, end=end_date, limit=limit_rows).df
column_names = [(tickers[0], x) for x in prices_ticker_1_df.columns]
prices_ticker_1_df.columns = pd.MultiIndex.from_tuples(column_names)
display(prices_ticker_1_df.head())
display(prices_ticker_1_df.tail())
```

```python
prices_ticker_2_df = alpaca.get_bars(tickers[1], timeframe ,start=start_date, end=end_date, limit=limit_rows).df
column_names = [(tickers[1], x) for x in prices_ticker_2_df.columns]
prices_ticker_2_df.columns = pd.MultiIndex.from_tuples(column_names)
display(prices_ticker_2_df.head())
display(prices_ticker_2_df.tail())
```

```python
price_hist_stock_bond_df = pd.merge(prices_ticker_1_df, prices_ticker_2_df, how = 'inner', left_index=True, right_index=True)
display(price_hist_stock_bond_df.head())
display(price_hist_stock_bond_df.tail())
```

```python
# Configure the Monte Carlo simulation to forecast 30 years cumulative returns
# The weights should be split 40% to AGG and 60% to SPY.
# Run 500 samples.
mc_60_40 = MCSimulation(
    portfolio_data=price_hist_stock_bond_df,
    weights=[.40,.60],
    num_simulation=500,
    num_trading_days=252*30)

# Review the simulation input data
mc_60_40.portfolio_data.head()
```

```python
# Run the Monte Carlo simulation to forecast 30 years cumulative returns
mc_60_40.calc_cumulative_return()
```

```python
# Visualize the 30-year Monte Carlo simulation by creating an
# overlay line plot
mc_60_40.plot_simulation()
```
